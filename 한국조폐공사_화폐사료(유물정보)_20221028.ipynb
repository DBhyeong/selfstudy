```python
from urllib.request import urlopen
from urllib.parse import urlencode, unquote, quote_plus
import urllib
import requests
import json
import pandas as pd
from datetime import datetime,timedelta
import warnings
import xmltodict # 결과가 xml 형식으로 반환된다. 이것을 dict 로 바꿔주는 라이브러리다

#한국조폐공사_화폐사료(유물정보)

#화폐사료(유물정보)

warnings.filterwarnings('ignore')

url = 'https://www.komsco.com/openAPI/getData.jsp'

req = requests.get(url, verify=False)

rawdata = req.text

#데이터에 <, >, " 같은 특수기호들이 &lt; &gt; &quot; 같은 식으로 저장되어 있어서 이 문자들을 특수기호로 다시 바꿔준
cleaned_data = rawdata.replace("&lt;", "<").replace("&gt;", ">").replace("&quot;", "\"") 


#ValueError: Expecting , delimiter: line 1 column 오류 해결

cleaned_data = cleaned_data.replace('"{', '{').replace('}"', '}').replace('\\"', '\"')

json_data = json.loads(cleaned_data)

json_data_2 = json_data['data']

json_komsco_data = pd.DataFrame(json_data_2)
json_komsco_data.head()
```
